```python
import kagglehub
import pandas as pd

# Descargar el dataset de Kaggle
path = kagglehub.dataset_download("fedesoriano/company-bankruptcy-prediction")

# Cargar los datos
data = pd.read_csv(path + "/data.csv")

# Mostrar las primeras filas para revisar
data.head()
```

```python
if "Unnamed: 0" in data.columns:
    data = data.drop(["Unnamed: 0"], axis=1)
```

```python
import json

data_dict = data.to_dict(orient="list")
data_json = json.dumps({"data": [data_dict]})

print("Ejemplo de JSON enviado:")
print(data_json[:500])  # Muestra los primeros 500 caracteres para verificar
```

```python
import requests

# Cargar la URI del servicio desde el archivo JSON
with open("uri.json", "r") as f:
    scoring_uri = json.load(f)["URI"][0]

# Definir headers
headers = {"Content-Type": "application/json"}

# Enviar datos al endpoint
response = requests.post(scoring_uri, data=data_json, headers=headers)

# Verificar respuesta
if response.status_code == 200:
    result = response.json()
    print("Predicciones recibidas:", result)
else:
    print(f"Error en la solicitud: {response.text}")
```
